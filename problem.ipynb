```python
import matplotlib.pyplot as plt
```

```python
def nr_of_grains(days):
    total = 1
    amounts = []
    amounts.append(total)
    
    for i in range(0, days):
        total = 2 * total
        amounts.append(total)

    return amounts   
```

```python
days = int(input('Insert the number of days: '))
amounts = nr_of_grains(days)
```

```python
print(f'the amount of grains after {days} days is {amounts[-1]}')
```

```python
plt.title('Number of grains of rice on the 100th day')
```

```python
plt.xlabel('Days')
plt.ylabel('Nr of grains')
```

```python
plt.plot(amounts, color='black', marker='o', linewidth=2, markersize=5)
```
